```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Tus datos de ejemplo de accesos por IP
ip_count = {'192.168.1.1': 10, '192.168.1.2': 15, '192.168.1.3': 5}  # Reemplaza con tus datos

# Extraer los datos para el gráfico
ips = list(ip_count.keys())
counts = list(ip_count.values())
x = np.arange(len(ips))  # Índices para cada IP

# Colores para cada barra
colors = ['red', 'pink', 'purple']  # Puedes agregar más colores según sea necesario

# Crear gráfico de barras
fig, ax = plt.subplots()

ax.bar(x, counts, color=colors, width=0.5, edgecolor="black", linewidth=0.7)
ax.set_xticks(x)
ax.set_xticklabels(ips, rotation=45)
ax.set_xlabel("IPs")
ax.set_ylabel("Número de accesos")
ax.set_title("Accesos por IP")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# Tus datos de códigos HTTP
code_count = {'200': 50, '500': 10}  # Reemplaza con tus datos reales

# Preparar los datos para el gráfico
labels = list(code_count.keys())  # Etiquetas para los códigos HTTP, ej: '200', '500'
sizes = list(code_count.values())  # Cantidades de cada código

# Cambiar los colores (aquí puedes ajustar el colormap o definir colores específicos)
colors = ['pink', 'purple']  # Colores personalizados para cada segmento

# Crear el gráfico circular
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "black"}, autopct='%1.1f%%', frame=True)

# Ajustes de visualización
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

## El gráfico de barras muestra la cantidad de veces que cada IP ingresó a la página,presentando a su vez cuál ingresó una mayor cantidad de veces. 
## El gráfico circular muestra la proporción de cada código HTTP presente en el conjunto de datos.
